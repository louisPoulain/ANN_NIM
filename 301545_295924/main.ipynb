# Q-Learning agent

For this part, all relevant results have been stored in a pickle dictionnary. It can be open using the following function load_dict.  
Its entries are as follows:  
- 'Q1': dictionnary  
    - 'samples_5': dictionnary  
        - 'Rewards': list containing the rewards at each step

- 'Q2': dictionnary
    - same entries as 'Q1'
- 'Q3': dictionnary
    - 'samples_5': dictionnary

        - 'mrand': dictionnary
            - the entries are the values used for $n_*$ and each contains the final value of $M_\text{rand}$
        - 'Training_Opt':
            - the entries are the values used for $n_*$ and each contains the time (in s) to achieve 80% of $M_\text{opt}$
        - 'Training_Rand':
            - the entries are the values used for $n_*$ and each contains the time (in s) to achieve 80% of $M_\text{rand}$
- 'Q4': dictionnary
    - 'samples_5': dictionnary
        - 'mopt_final': dictionnary
            - the entries are the values used for $\varepsilon_\text{opt}$ ('0', '0.01', '0.05', '0.1', '0.8') and each contains the final value of $M_\text{opt}$
        - 'mrand_final': dictionnary
            - the entries are the values used for $\varepsilon_\text{opt}$ and each contains the final value of $M_\text{rand}$
        - 'Training_Opt':
            - the entries are the values used for $\varepsilon_\text{opt}$ and each contains the time (in s) to achieve 80% of $M_\text{opt}$
        - 'Training_Rand':
            - the entries are the values used for $\varepsilon_\text{opt}$ and each contains the time (in s) to achieve 80% of $M_\text{rand}$
- 'Q7': dictionnary  
    - 'samples_5': dictionnary
        - 'mopt_final': dictionnary
            - the entries are the values used for $\varepsilon$ ('0', '0.05', '0.1', '0.2', '0.8') and each contains the final value of $M_\text{opt}$
        - 'mrand_final': dictionnary
            - the entries are the values used for $\varepsilon$ and each contains the final value of $M_\text{rand}$
        - 'Training_Opt':
            - the entries are the values used for $\varepsilon$ and each contains the time (in s) to achieve 80% of $M_\text{opt}$
        - 'Training_Rand':
            - the entries are the values used for $\varepsilon$ and each contains the time (in s) to achieve 80% of $M_\text{rand}$
- 'Q8': dictionnary 
    - 'samples_5': dictionnary
        - 'mopt_final': dictionnary
            - the entries are the values used for $n_*$ ('1', '1000', '5000', '10000', '20000') and each contains the final value of $M_\text{opt}$
        - 'mrand_final': dictionnary
            - the entries are the values used for $n_*$ and each contains the final value of $M_\text{rand}$
        - 'qvals_final': dictionnary
            - the entries are the values used for $n_*$. Each entry is itself a dictionnay containing all the Q-values of the player after the training.
